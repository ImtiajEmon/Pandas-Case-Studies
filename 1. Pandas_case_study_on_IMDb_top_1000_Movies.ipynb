**16. Write a function which takes genre as argument and return average gross and total gross.**

```python
def give_gross(genre):
    mask = movies['Genre'] == genre
    avg_gross = movies[mask]['Gross'].mean()
    total_gross = movies[mask]['Gross'].sum()

    return avg_gross, total_gross

genre = "Action"
avg, total = give_gross(genre)

print(f"Genre: {genre}\nAverage Gross: {avg}\nTotal Gross: {total}")
```

**20. Find the highest rated movie of each genre.**

```python
# Approach - 1

series_of_max_rating = movies.groupby('Genre')['IMDB_Rating'].max()
genres = list(series_of_max_rating.index)
max_ratings = list(series_of_max_rating.values)

ans = pd.DataFrame()

for genre, max_rating in zip(genres, max_ratings):
    mask = ((movies['Genre'] == genre) & (movies['IMDB_Rating'] == max_rating))
    ans = pd.concat([ans, movies[mask]], ignore_index = True)

ans[['Genre', 'Series_Title', 'IMDB_Rating']]
```

```python
# Approach - 2: Looping over Groups

genres = movies.groupby('Genre')

for genre, df in genres:
    mask = df["IMDB_Rating"] == df['IMDB_Rating'].max()
    print(genre, ": ", list(df[mask]["Series_Title"]))
```

```python
# Approach - 3: Using Apply

def max_rated(genre):
    global ans
    mask = genre["IMDB_Rating"] == genre['IMDB_Rating'].max()
    ans = pd.concat([ans, genre[mask]], ignore_index = True)


genres = movies.groupby('Genre')
ans = pd.DataFrame()
genres.apply(max_rated)

ans[['Genre', 'Series_Title', 'IMDB_Rating']]
```

**21. Write a function which takes genre as aurgument and return highest rated movie of that genre.**

```python
def max_rated_movie(genre):
    mask = movies["Genre"] == genre
    df = movies[mask]

    return df[df["IMDB_Rating"] == df["IMDB_Rating"].max()]

max_rated_movie("Action")
```

**22. Find number of movies starting with A for each group.**

```python
def movie_count(genre):
    return genre["Series_Title"].str.startswith('A').sum()

genres = movies.groupby('Genre')
genres.apply(movie_count)
```

**23. Find ranking of each movie in each genre according to IMDB score. Rank 1 goes to the highest rated movie, 2 goes to the 2nd highest rated movie and so on.**

```python
def find_rank(genre):
    genre["Rank"] = genre["IMDB_Rating"].rank(ascending = False)
    return genre

genres = movies.groupby('Genre')
genres.apply(find_rank)
```
